# Recommendation Model

```python
data_final = data.astype(bool).astype(int)
```

```python
data_final.head()
```

```python
freq_movies = apriori(data_final, min_support=0.0004,use_colnames = 1).sort_values('support',ascending = False)
freq_movies
```

```python
# Association Rules & Info
df_ar = association_rules(freq_movies, metric = "confidence", min_threshold = 0.0004)
df_ar
```

```python
df_ar[(df_ar['antecedent support'] > 0.001) & (df_ar['consequent support'] > 0.001) & 
      (df_ar['lift'] > 1)].sort_values(["lift", "confidence"], ascending = [False, False])
```
